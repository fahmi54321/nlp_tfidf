```python
import pandas as pd
import numpy as np
import nltk

from nltk import word_tokenize
```

```python
nltk.download('punkt')
```

```python
df = pd.read_csv('bbc_text_cls.csv')
```

```python
df.head()
```

```python
# populate word2idx
# convert documents into sequences of ints / ids / indices

idx = 0
word2idx = {}
tokenized_docs = []
for doc in df['text']:
  words = word_tokenize(doc.lower())
  doc_as_int = []
  for word in words:
    if word not in word2idx:
      word2idx[word] = idx
      idx += 1

    # save for later
    doc_as_int.append(word2idx[word])
  tokenized_docs.append(doc_as_int)
```

```python
# reverse mapping
# if you do it smarter you can store it as a list
idx2word = {v: k for k, v in word2idx.items()}
print(idx2word)
```

```python
# number of documents
N = len(df['text'])
print(N)
```

```python
# number of words
V = len(word2idx)
print(V)
```

```python
# instatiate term-frequency matrix
# note: could have also used count vectorizer
tf = np.zeros((N,V))
print(tf)
```

```python
# populate term-frequency counts
for i, doc_as_int in enumerate(tokenized_docs):
  for j in doc_as_int:
    tf[i,j] += 1

print(tf)
```

```python
# compute IDF
document_freq = np.sum(tf > 0, axis = 0) # document frequency (shape = (V,))
idf = np.log(N / document_freq)

print(idf)
```

```python
# compute TF-IDF
tf_idf = tf * idf
print(tf_idf)
```

```python
np.random.seed(123)
```

```python
# pick a random document, show the top 5 terms (in terms of tf_idf score)
i = np.random.choice(N)
row = df.iloc[i]
print('Label: ',row['labels'])
print('Text: ',row['text'].split("\n", 1)[0])
print("top 5 terms:")

scores = tf_idf[i]
indices = (-scores).argsort()

for j in indices[:5]:
  print(idx2word[j])
```
